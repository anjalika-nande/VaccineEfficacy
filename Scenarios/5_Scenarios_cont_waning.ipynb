# Exploring Scenarios - Continuous Vaccination With Waning Vaccine Efficacy

Here we investigate the effect of continuous vaccination with waning efficacy. We explore the same 9 scenarios with the following parameter values: $\gamma = 1/4$, $\sigma=1/4$, $w=0.0075$, $v=0.01$, and the following initial conditions: $S(0) = 0.98$, $E(0) = 0.01$, $I(0) = 0.01$, $R(0) = 0$, $V_S = 0$, and $V_R = 0$. 

Similar to previous notebooks, we plot the difference in percentage reducction of total recovered population of the leaky and all-or-nothing vaccines with respect to that of without vaccination, measured at 30 days, 60 days, 100 days, and 200 days after vaccination. 

Note that we assume that no reinfection occurs.

```python
%load_ext blackcellmagic
```

```python
### import modules
import numpy as np
import matplotlib.pyplot as plt
from utils import run_scenarios
from plot import plot_scenarios
import warnings

warnings.filterwarnings("ignore")
```

## Measured 30 Days After Vaccination

```python
# define parameter values and initial conditions
sigma = 1 / 4
gamma = 1 / 4
w = 0.0075
t = 200
covs = np.array([0.5, 0.75, 1.0])
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.linspace(1.0, 3.0, 601)
epss = np.linspace(0.01, 1, 201)

# run scenarios varying R0 values and vaccine efficacy
size0_m30_cont_w = run_scenarios(
    y0=y0,
    t=t,
    size=0,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=w,
    epss=epss,
    measured=30,
    vax_mode='cont',
    v=0.01,
)
size10_m30_cont_w = run_scenarios(
    y0=y0,
    t=t,
    size=0.1,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=w,
    epss=epss,
    measured=30,
    vax_mode='cont',
    v=0.01,
)
size25_m30_cont_w = run_scenarios(
    y0=y0,
    t=t,
    size=0.25,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=w,
    epss=epss,
    measured=30,
    vax_mode='cont',
    v=0.01,
)

# plot difference in percentage reduction
plt.rcParams["pdf.fonttype"] = 42
fig_3d = plot_scenarios(
    size0_m30_cont_w, size10_m30_cont_w, size25_m30_cont_w, R0s, epss, dim=3, mask_hit=False
)
plt.savefig("diffPerc_m30_cont_3d_waning.pdf", format="pdf", bbox_inches="tight")
fig_2d = plot_scenarios(
    size0_m30_cont_w, size10_m30_cont_w, size25_m30_cont_w, R0s, epss, dim=2, mask_hit=False
)
plt.savefig("diffPerc_m30_cont_2d_waning.pdf", format="pdf", bbox_inches="tight")

# store dataframes
%store size0_m30_cont_w
%store size10_m30_cont_w
%store size25_m30_cont_w
```
