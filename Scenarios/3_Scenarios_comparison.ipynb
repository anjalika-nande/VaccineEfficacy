# Comparing With and Without Waning

Here we compare the results from running the modified SEIR model with and without waning protection of the vaccines. We compare the difference in the percentage reduction of the total recovered population between the two vaccine failure modes for when the vaccines are 50% and 80% effective, calculated 30 days after vaccination, 60 days after vaccination, 100 days after vaccination, and 200 days after vaccination. The blank spaces correspond to $R_0$ values for which $f_V^* < 0$ or if the vaccine is never administered because the epidemic ends before reaching the target recovered population size.

## 50% efficacy
### Measured 30 Days After Vaccination

```python
%store -r size0_m30
%store -r size10_m30
%store -r size25_m30

%store -r size0_m30_w
%store -r size10_m30_w
%store -r size25_m30_w

R0s = np.linspace(1.0, 3.0, 601)

plt.rcParams["pdf.fonttype"] = 42
fig = plot_comparison(
    size0_m30,
    size10_m30,
    size25_m30,
    size0_m30_w,
    size10_m30_w,
    size25_m30_w,
    R0s,
    eps=0.5,
    tol=0.0025,
)
plt.savefig("comparison_eps50_m30.pdf", format="pdf", bbox_inches="tight")
```
