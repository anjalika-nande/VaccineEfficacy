# Exploring Scenarios - With Waning Vaccine Efficacy

In order to investigate the impact of waning protection of vaccines, we introduce a constant vaccine waning rate $w$ and explore the same 9 scenarios ($t_V$: $R=0, R=0.10, R=0.25$, $f_V$: 50%, 75%, 100% of $(1-f_R)$) with $R_0$ values varying from 1.0 to 3.0 and vaccine efficacy varying from 1% to 100% effective. 

Similar to [1_Scenarios](./1_Scenarios.ipynb), we then calculate the percentage reduction of total recovered population of the leaky and all-or-nothing vaccines with respect to that of without vaccination for each simulation run with varying $R_0$ and vaccine efficay for each scenario, and the plots below show the log-scaled differences. The colorbar to the right shows the true values. Furthermore, the blank spaces in the plots show where the critical vaccination threshold $f_V^*$ is below 0 or when the recovered population never reaches the target population size (the vaccine is never administered). The red lines divide regimes for which $f_V \leq f_V^*$ (above red line) vs. $f_V > f_V^*$ (below red line). The difference is measured at 30 days after vaccination, 60 days after vaccination, 100 days after vaccination, and the end of the modeled pandemic (at 200 days). 

Note that we assume the vaccine campaigns are instantaneous and that no reinfection occurs.

The simulations are run with the following parameter values: $\gamma = 1/4$, $\sigma=1/4$, $w=0.0075$, and the following initial conditions: $S(0) = 0.98$, $E(0) = 0.01$, $I(0) = 0.01$, $R(0) = 0$, $V_S = 0$, and $V_R = 0$. The waning rate $w$ is set such that half of the vaccinated population loses its immunity by 3 months.

For more details on the scenarios and the equations, see [1_Scenarios](./1_Scenarios.ipynb).

For more details on the model, see [Models](../Models/Models.ipynb).

```python
### import modules
import numpy as np
import matplotlib.pyplot as plt
from utils import run_scenarios
from plot import plot_scenarios
import warnings

warnings.filterwarnings("ignore")
```

## Measured 30 Days After Vaccination

```python
# define parameter values and initial conditions
sigma = 1 / 4
gamma = 1 / 4
w = 0.0075
t = 200
covs = np.array([0.5, 0.75, 1.0])
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.linspace(1.0, 3.0, 601)
epss = np.linspace(0.01, 1, 201)

# run scenarios varying R0 values and vaccine efficacy
size0_m30_w = run_scenarios(
    y0=y0,
    t=t,
    size=0,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=w,
    epss=epss,
    measured=30,
)
size10_m30_w = run_scenarios(
    y0=y0,
    t=t,
    size=0.1,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=w,
    epss=epss,
    measured=30,
)
size25_m30_w = run_scenarios(
    y0=y0,
    t=t,
    size=0.25,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=w,
    epss=epss,
    measured=30,
)

# plot difference in percentage reduction
plt.rcParams["pdf.fonttype"] = 42
fig_3d = plot_scenarios(
    size0_m30_w, size10_m30_w, size25_m30_w, R0s, epss, dim=3, mask_hit=False
)
plt.savefig("diffPerc_m30_3d_waning.pdf", format="pdf", bbox_inches="tight")
fig_2d = plot_scenarios(
    size0_m30_w, size10_m30_w, size25_m30_w, R0s, epss, dim=2, mask_hit=False
)
plt.savefig("diffPerc_m30_2d_waning.pdf", format="pdf", bbox_inches="tight")

# store dataframes
%store size0_m30_w
%store size10_m30_w
%store size25_m30_w
```
