# Exploring Scenarios - Continuous Vaccination

Here we investigate the effect of continuous vaccination and the vaccination rate on the two modes of vaccine failure. We run the same scenarios as [1_Scenarios](./1_Scenarios.ipynb), with the same parameter values: $\gamma = 1/4$, $\sigma=1/4$, $w=0$, and initial conditions: $S(0) = 0.98$, $E(0) = 0.01$, $I(0) = 0.01$, $R(0) = 0$, $V_S = 0$, and $V_R = 0$. We vary the vaccination rate $v=0.01, 0.03, 0.05$ for the following simulations, such that the maximum vaccination rate is 1%, 3%, or 5% of the susceptible population per day. Vaccine coverage is assumed to be 75%.

Similar to previous notebooks, we plot the difference in percentage reduction of total recovered population of the leaky and all-or-nothing vaccines with respect to that of without vaccination focusing on 200 days after vaccination. 

Note that we assume that no reinfection occurs, and vaccine efficacy doesn't wane overtime.

## Measured 200 Days After Vaccination (End of Pandemic)

```python
# define parameter values and initial conditions
sigma = 1 / 4
gamma = 1 / 4
t = 200
covs = np.array([0.75])
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.linspace(1.0, 3.0, 601)
epss = np.linspace(0.01, 1, 201)
```

### 1% vaccination rate

```python
vmax = 0.01

# run scenarios varying R0 values and vaccine efficacy
size0_m200_cont_01 = run_scenarios(
    y0=y0,
    t=t,
    size=0,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=0,
    epss=epss,
    measured=200,
    vax_mode='cont',
    v=vmax,
)
size10_m200_cont_01 = run_scenarios(
    y0=y0,
    t=t,
    size=0.1,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=0,
    epss=epss,
    measured=200,
    vax_mode='cont',
    v=vmax,
)
size25_m200_cont_01 = run_scenarios(
    y0=y0,
    t=t,
    size=0.25,
    covs=covs,
    R0s=R0s,
    sigma=sigma,
    gamma=gamma,
    w=0,
    epss=epss,
    measured=200,
    vax_mode='cont',
    v=vmax,
)

# Save results

size25_m200_cont_01['Vmax'] = vmax
size10_m200_cont_01['Vmax'] = vmax
size0_m200_cont_01['Vmax'] = vmax

size25_m200_cont_01.to_csv('size25_m200_cont_01.csv', index=False)
size10_m200_cont_01.to_csv('size10_m200_cont_01.csv', index=False)
size0_m200_cont_01.to_csv('size0_m200_cont_01.csv', index=False)
```
