# Vaccinating Fixed Proportion of Susceptible Population

Here we vaccinate a fixed proportion of the remaining susceptible population at the time of vaccination (50%, 75%, 100% of S). We have three scenarios for the timing of vaccination: when R reaches 0%, 10%, and 25% of the total population. We compare the difference in percentage reduction of the recovered population of the two vaccine failure modes compared to that of without vaccination (percentage reduction of all-or-nothing vaccine - percentage reduction of leaky vaccine). The difference is measured 30 days, 60 days, 100 days, and 200 days after vaccination (end of epidemic).

The blank spaces in the plots show where $f_V^* < 0$ or when the recovered population never reaches the target population size (the vaccine is never administered). The red lines divide regimes for which $f_V \leq f_V^*$ vs. $f_V > f_V^*$ (plot the largest vaccine efficacy value for each $R_0$ for which $f_V \leq f_V^*$); values above the red line are values for which $f_V > f_V^*$ (the vaccinated population is greater than the critical vaccination threshold), and values below the red line are values for which $f_V \leq f_V^*$.

$f_V^*$ is bounded by 0 and the initial susceptible population (0.98). Any $f_V^*$ values that are below 0 are removed from the calculation, and any $f_V^*$ values that are greater than the initial susceptible population are set equal to $S(0)$, with the understanding that if $f_V^* < S(0)$, $f_V$ is always less than or equal to $f_V^*$, since $0 \leq f_V \leq S(0)$.

Note that the vaccine is administered instantaneously.

## Measured 30 Days After Vaccination

```python
# define parameter values and initial conditions
sigma = 1 / 4
gamma = 1 / 4
t = 200
covs = np.array([0.5, 0.75, 1.0])
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.linspace(1.0, 3.0, 601)
epss = np.linspace(0.01, 1, 201)

# run scenarios varying R0 values and vaccine efficacy
epi0_m30 = run_scenarios_Spop(y0, t, 0, covs, R0s, sigma, gamma, epss, measured=30)
epi10_m30 = run_scenarios_Spop(y0, t, 0.1, covs, R0s, sigma, gamma, epss, measured=30)
epi25_m30 = run_scenarios_Spop(y0, t, 0.25, covs, R0s, sigma, gamma, epss, measured=30)

# plot difference in total effectiveness for each scenario
fig_3d = plot_scenarios_Spop(
    epi0_m30, epi10_m30, epi25_m30, R0s, epss, dim=3, mask_hit=False
)
fig_2d = plot_scenarios_Spop(
    epi0_m30, epi10_m30, epi25_m30, R0s, epss, dim=2, mask_hit=False
)
```
