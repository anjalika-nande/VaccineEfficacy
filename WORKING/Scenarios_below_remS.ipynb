# Below $f_V^*$, Vaccinating the Remaining Susceptible Population 

Here we investigate two scenarios: one where $f_V < f_V^*$ ($f_V = 0.8 f_V^*$), and one where $f_V = S(t_V)$, $t_V$ being the timing of vaccination. We investigate three $t_V$ values for each scenario: when the epidemic size reaches 0%, 10%, and 25%, where epidemic size is measured by the recovered population. We assume instantaneous vaccination.

The difference in percentage reduction of the two vaccine failure modes are measured at 30, 60, 100, and 200 days after vaccination (end of epidemic).

```python
%load_ext blackcellmagic
```

```python
import numpy as np
from utils import run_scenarios_below_remS
from plot import plot_scenarios_below_remS
import warnings

warnings.filterwarnings("ignore")
```

The following plots show the difference in the percentage reduction of the recovered population between the two vaccine failure modes. The difference is calculated by subtracting the percentage reduction of the leaky vaccine from that of the all-or-nothing vaccine (the all-or-nothing vaccine always performs better). The blank spaces in the plots show where $f_V^* < 0$ or when the recovered population never reaches the target population size (the vaccine is never administered). The red lines divide regimes for which $f_V \leq f_V^* vs. $f_V > f_V^*$ (plot the largest vaccine efficacy value for each $R_0$ for which $f_V \leq f_V^*$); values above the red line are values for which $f_V > f_V^*$ (the vaccinated population is greater than the critical vaccination threshold), and values below the red line are values for which $f_V \leq f_V^*$. 

$f_V^*$ is bounded by 0 and the initial susceptible population (0.98). Any $f_V^*$ values that are below 0 are removed from the calculation, and any $f_V^*$ values that are greater than the initial susceptible population are set equal to $S(0)$, with the understanding that if $f_V^* < S(0)$, $f_V$ is always less than or equal to $f_V^*$, since $0 \leq f_V \leq S(0)$. 

Note that the vaccine is administered instantaneously, and for the scenario for which the vaccine is administered before the epidemic (vaccination at 0% epidemic size), we vaccinate the entire susceptible population (MAYBE CHANGE).

## Measured 30 Days After Vaccination

```python
sigma = 1 / 4
gamma = 1 / 4
t = 200
covs = ["Below fc", "Remaining S"]
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.linspace(1.0, 3.0, 601)
epss = np.linspace(0.01, 1, 201)

# run scenarios varying R0 values and vaccine efficacy
epi0_m30 = run_scenarios_below_remS(
    y0, t, 0, covs, R0s, sigma, gamma, epss, measured=30
)
epi10_m30 = run_scenarios_below_remS(
    y0, t, 0.1, covs, R0s, sigma, gamma, epss, measured=30
)
epi25_m30 = run_scenarios_below_remS(
    y0, t, 0.25, covs, R0s, sigma, gamma, epss, measured=30
)

print(epi0_m30.isnull().any().any())
print(epi10_m30.isnull().any().any())
print(epi25_m30.isnull().any().any())

# plot difference in total effectiveness for each scenario
fig_3d = plot_scenarios_below_remS(
    epi0_m30, epi10_m30, epi25_m30, R0s, epss, dim=3, mask_hit=False
)
fig_2d = plot_scenarios_below_remS(
    epi0_m30, epi10_m30, epi25_m30, R0s, epss, dim=2, mask_hit=False
)
```
