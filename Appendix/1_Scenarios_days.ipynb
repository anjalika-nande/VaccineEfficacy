# Exploring Scenarios - Based on Number of Days Since Beginning of Epidemic

Here, we explore different regimes where vaccination timing is determined by the number of days that has passed since the epidemic begins, and vaccination level depends on critical vaccination level calculated for each $R_0$ and vaccine efficacy value. The 9 scenarios investigated are:
- Pre-transmission vaccination, covering $f_{V, \text{ Below}}$ of population
- Pre-transmission vaccination, covering $f_{V, \text{ Slightly Above}}$ of population
- Pre-transmission vaccination, covering $f_{V, \text{ Above}}$ of population
- 10 days post-transmission vaccination, covering $f_{V, \text{ Below}}$ of population
- 10 days post-transmission vaccination, covering $f_{V, \text{ Slightly Above}}$ of population
- 10 days post-transmission vaccination, covering $f_{V, \text{ Above}}$ of population
- 30 days post-transmission vaccination, covering $f_{V, \text{ Below}}$ of population
- 30 days post-transmission vaccination, covering $f_{V, \text{ Slightly Above}}$ of population
- 10 days post-transmission vaccination, covering $f_{V, \text{ Above}}$ of population

For each scenario, the $R_0$ value is varied between 1.0 and 3.0, and vaccine efficacy is varied between 0.01 and 1. We assume the vaccine campaigns are instantaneous, no reinfection ocurrs, and vaccine efficacy does not wane overtime. For more information on parameter values, initial conditions, and plots generated, see [1_Scenarios](../Scenarios/1_Scenarios.ipynb').

## Measured 30 Days After Vaccination

```python
# define parameter values and initial conditions
sigma = 1 / 4
gamma = 1 / 4
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.arange(1.0, 3.0, 0.01)
epss = np.arange(0.01, 1.0, 0.01)

pre_m30 = run_scenarios_days(
    y0=y0, t=30, tv=-1, R0s=R0s, sigma=sigma, gamma=gamma, w=0, epss=epss
)
post10_m30 = run_scenarios_days(
    y0=y0, t=40, tv=10, R0s=R0s, sigma=sigma, gamma=gamma, w=0, epss=epss
)
post30_m30 = run_scenarios_days(
    y0=y0, t=60, tv=30, R0s=R0s, sigma=sigma, gamma=gamma, w=0, epss=epss
)

# plot difference in percentage reduction
plt.rcParams["pdf.fonttype"] = 42
fig_3d = plot_scenarios_days(pre_m30, post10_m30, post30_m30, R0s, epss, dim=3)
plt.savefig("days_diffPerc_m30_3d.pdf", format="pdf", bbox_inches="tight")
fig_2d = plot_scenarios_days(pre_m30, post10_m30, post30_m30, R0s, epss, dim=2)
plt.savefig("days_diffPerc_m30_2d.pdf", format="pdf", bbox_inches="tight")
```
