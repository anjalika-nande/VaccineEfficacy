# Exploring Scenarios - Based on Number of Days Since Beginning of Epidemic

Here, we explore the same 9 scenarios as [1_Scenarios_days](./1_Scenarios_days.ipynb) but with vaccine efficacy waning at a constant rate over time. 

The simulations are run with the following parameter values: $\gamma = 1/4$, $\sigma=1/4$, $w=0.0075$, and the following initial conditions: $S(0) = 0.98$, $E(0) = 0.01$, $I(0) = 0.01$, $R(0) = 0$, $V_S = 0$, and $V_R = 0$. The waning rate $w$ is set such that half of the vaccinated population loses its immunity by 3 months.

For each scenario, the $R_0$ value is varied between 1.0 and 3.0, and vaccine efficacy is varied between 0.01 and 1. We assume the vaccine campaigns are instantaneous and that no reinfection ocurrs. For more information on the scenarios and plots generated, see [1_Scenarios](../Scenarios/1_Scenarios.ipynb').

## Measured 30 Days After Vaccination

```python
# define parameter values and initial conditions
sigma = 1 / 4
gamma = 1 / 4
w = 0.075
s0 = 0.98
e0 = 0.01
i0 = 0.01
r0 = 0
y0 = [s0, e0, i0, r0]
R0s = np.arange(1.0, 3.0, 0.01)
epss = np.arange(0.01, 1.0, 0.01)

pre_m30_w = run_scenarios_days(
    y0=y0, t=30, tv=-1, R0s=R0s, sigma=sigma, gamma=gamma, w=w, epss=epss
)
post10_m30_w = run_scenarios_days(
    y0=y0, t=40, tv=10, R0s=R0s, sigma=sigma, gamma=gamma, w=w, epss=epss
)
post30_m30_w = run_scenarios_days(
    y0=y0, t=60, tv=30, R0s=R0s, sigma=sigma, gamma=gamma, w=w, epss=epss
)

# plot difference in percentage reduction
plt.rcParams["pdf.fonttype"] = 42
fig_3d = plot_scenarios_days(pre_m30_w, post10_m30_w, post30_m30_w, R0s, epss, dim=3)
plt.savefig("days_diffPerc_m30_3d_waning.pdf", format="pdf", bbox_inches="tight")
fig_2d = plot_scenarios_days(pre_m30_w, post10_m30_w, post30_m30_w, R0s, epss, dim=2)
plt.savefig("days_diffPerc_m30_2d_waning.pdf", format="pdf", bbox_inches="tight")
```
